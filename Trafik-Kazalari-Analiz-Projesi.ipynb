```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("accidents_2012_to_2014.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
print(df["Accident_Index"].unique())
print(df["Accident_Index"].value_counts())
```

```python
df2= df.loc[:,["Accident_Severity","Number_of_Casualties","Day_of_Week","Date","Time","Local_Authority_(District)"
                         ,"Local_Authority_(Highway)","Road_Type","Speed_limit","Light_Conditions","Weather_Conditions",
                        "Road_Surface_Conditions","Year"]]
df.head()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna()
df2.isnull().sum()
```

```python
df2.Year = df2.Year.astype(str)
accidentNumber = df2.groupby("Year") #iki yÄ±l iÃ§inde grupladÄ±k
sum_accidentNumber=pd.DataFrame(accidentNumber.size().sort_values(ascending=False),columns=["veri sayÄ±sÄ±"])
sum_accidentNumber.head()
```

```python
sns.barplot(y = sum_accidentNumber["veri sayÄ±sÄ±"] , x = sum_accidentNumber.index)
plt.title("YÄ±llara gÃ¶re kaza sayÄ±sÄ±")
```

```python
df.Speed_limit = df.Speed_limit.astype(str)
limit_veri_sayi = df.groupby("Speed_limit")
toplam = pd.DataFrame(limit_veri_sayi.size().sort_values(ascending=False),columns=["toplam kaza"])
sns.barplot(y = toplam["toplam kaza"] , x = toplam.index)
plt.title("HÄ±z Limitine gÃ¶re Kaza sayÄ±sÄ±")
```

```python
df.Road_Type = df.Road_Type.astype(str)
tip_veri_sayi = df.groupby("Road_Type")
toplam = pd.DataFrame(tip_veri_sayi.size().sort_values(ascending=False),columns=["toplam kaza"])
sns.barplot(x = toplam["toplam kaza"] , y = toplam.index)
plt.title("Yol Tipine gÃ¶re Kaza sayÄ±sÄ±")
```

```python
df.Day_of_Week = df.Day_of_Week.astype(str)
gun_veri_sayi = df.groupby("Day_of_Week")
toplam = pd.DataFrame(gun_veri_sayi.size().sort_values(ascending=False),columns=["toplam kaza"])
sns.barplot(x = toplam["toplam kaza"] , y = toplam.index)
plt.title("HaftanÄ±n gÃ¼nlerine gÃ¶re Kaza sayÄ±sÄ±")
```
